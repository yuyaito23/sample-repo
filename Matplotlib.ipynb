```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline 
#magic command alllows to use a graph on this note
```

```python
#plt.plot()

x = [1,2,3,4,5]
#y = [1,4,9,16,25]
y = np.array(x) ** 2
plt.plot(x, y)
```

```python
#シグモイド関数　
#if x is minus, x gets close to 0. if x is plus, x gets close to 1.
x = np.linspace(-10, 10, 100)
z = 1 / (1 + np.exp(-x))
plt.plot(x, z)
```

```python
plt.plot(x, z, color='red', lw=3, ls='--', marker='o', markersize=5, markerfacecolor='yellow', markeredgecolor='blue', markeredgewidth=4, alpha=0.2) 
#'r', color='red', lw=3, linewidth=3, 
#ls='--', market='o', markersize=5, markerfacecolor='yellow', markeredgecolor='blue', markeredgewidth=4, alpha=0.2, alpha means transparency of this. 0 to 1, 0 is clear.
```

```python
# 付属情報
plt.plot(x, z, label='sigmoid')
plt.plot(x, z*2, label='sigmoid * 2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sigmoid(x)')
plt.legend(loc=4)
plt.xticks([1,2,3], rotation=20)
plt.yticks([0,0.5,1], rotation=20)
```

### **Seaborn**

```python
import seaborn as sns
%matplotlib inline 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/tmdb_5000_movies.csv')
sns.distplot(df['vote_count'])
```

```python
#sns.jointplot, pairplot
sns.jointplot(x='vote_count', y='vote_average', data=df, alpha=0.2)
```

```python
df = pd.read_csv('/content/tmdb_5000_movies.csv')
df = df[df['vote_count'] > 100]
sns.jointplot(x='vote_count', y='vote_average', data=df, kind='reg', joint_kws={'scatter_kws': {'alpha': 0.3}}) #kind='reg'
```

```python
sns.pairplot(df[['revenue','budget', 'vote_count', 'vote_average']])
```

```python
df['evaluation'] = df['vote_average'].apply(lambda x: 'high' if x > df["vote_average"].median() else 'low')
sns.pairplot(df[['revenue', 'budget', 'vote_count', 'vote_average', 'evaluation']], 
             hue='evaluation',
             plot_kws={'alpha':0.3},
             diag_kind='hist')
```

```python
#barplot
sns.barplot(x='evaluation', y='revenue', data=df, estimator=np.median)
```

```python
df['year'] = df['release_date'].apply(lambda x: x[:4])
```

```python
ax = sns.countplot(x='year', data=df, hue='evaluation') #axis object
ax.tick_params(rotation=90, labelsize=6)
```

```python
df['decade'] = df['release_date'].apply(lambda x: x[:3] + '0s')
```

```python
df
```

```python
sns.countplot(x='decade', data=df.sort_values('decade'), hue='evaluation')
```

```python
sns.boxplot(x='decade', y='vote_average', data=df.sort_values('decade'))
```

```python
sns.violinplot(x='decade', y='vote_average', data=df.sort_values('decade'))
```

```python
sns.swarmplot(x='decade', y='vote_average', data=df.sort_values('decade'))
```

```python
fig, ax = plt.subplots(figsize=(20, 5))
ax.tick_params(labelsize=20)
sns.swarmplot(ax=ax, x='decade', y='vote_average', data=df.sort_values('decade'))
```

```python
# Heatmap
# corr 相関係数 -1, 1の値で表示
df = pd.read_csv('/content/tmdb_5000_movies.csv')
df = df.loc[(df != 0).all(axis=1)]
df.corr()
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm') # annotaion
```

```python
df['year'] = df['release_date'].apply(lambda x: x[:4])
df['month'] = df['release_date'].apply(lambda x: x[5:7])
revenue_df = df.groupby(['year', 'month']).mean()[['revenue']].pivot_table(values='revenue', index='month', columns='year')
```

```python
sns.heatmap(revenue_df)
```

```python
# sns change style, context: 'paper', 'notebook', 'talk', 'poster'
#sns.set(context='poster')
sns.set_context('poster')
sns.displot(df['vote_average'])
```

```python
# style: 'darkgrid', 'whitegrid', 'dark', 'white', 'ticks'
sns.set(context='notebook', style='ticks')
sns.displot(df['vote_average'])
sns.despine()
```

```python
# platte: deep, muted, bright, pastel, dark, colorblind
sns.set(context='notebook', style='ticks', palette='bright')

df = pd.read_csv('/content/tmdb_5000_movies.csv')
df = df[~df['release_date'].isna()]
df['decade'] = df['release_date'].apply(lambda x: x[:3] + '0s')
df['evaluation'] = df['vote_average'].apply(lambda x: 'high' if x > df['vote_average'].median() else 'low')

sns.violinplot(x='decade', y='popularity', data=df.sort_values('decade'), palette='twilight')
```

```python
sns.distplot(df['vote_average'])
plt.title('Distribution of scores')
plt.xlabel('Evaluation Scores')
plt.savefig('saved_seaborn.png')
```
